```python
import pandas as pd
df=pd.read_csv('C:/Users/tneha/OneDrive/Desktop/Git_Hub/Datasets/california_housing_test.csv')
print(df.head(10))
print(df.tail(5))
```

```python
print(df.describe())
print(df.dtypes)
```

```python
filtered_df = df[df['median_income'] > 4]
print(filtered_df.head())
```

```python
df['rooms_per_household'] = df['total_rooms'] / df['households']
```

```python
grouped = df.groupby('housing_median_age')['median_house_value'].mean()
print("\nGrouped Mean House Value by Age:")
print(grouped.head())
```

```python
sorted_df = df.sort_values(by='median_house_value', ascending=False)
print("\nTop 5 Expensive Houses:")
print(sorted_df.head())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.to_csv("california_test_cleaned.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.histplot(df['median_income'], kde=True, bins=20)
plt.title("Distribution of Median Income")
plt.xlabel("Median Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='housing_median_age', y='median_house_value', data=df)
plt.title("House Value by Age")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_rooms', y='median_house_value', data=df)
plt.title("Total Rooms vs House Value")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['median_income', 'total_rooms', 'median_house_value']])
plt.suptitle("Pairwise Plot", y=1.02)
plt.show()
```
